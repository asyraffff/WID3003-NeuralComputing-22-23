```python
import numpy as np
import random

# Initialize inputs, bias, weights and initial prediction
inputs = np.array([[-1,-1], [-1,1], [1,-1], [1,1]])
# weights = np.array([0,0]) # randomize this later
weights = np.random.random([1,2])
print(f'initial inputs : {inputs}')
print(f'initial weights : {weights}')
print(f'len input : {len(inputs)}')

# AND settings
outputs = np.array([-1,-1,-1,1])
bias = -1 # from the answer in the slide. You can update the bias.
print(f'initial outputs : {outputs}')
print(f'initial bias : {bias}')

# OR settings
# outputs = np.array([-1,1,1,1])
# bias = 1

predict = np.array([0,0,0,0]) # initial prediction

error = sum ((outputs-predict)**2)

learning_rate = 0.01

def predict_fun(weights, input, bias):
    # summation function
    sum_func = np.dot(weights, input) + bias
    
    # activation function (sgn, signal function)
    if sum_func < 0:
        return -1
    elif sum_func == 0:
        return 0
    elif sum_func > 0:
        return 1
    
iter = 0

while error!=0: # in 'real dataset', this is not possible
    iter+=1
    for i in range(len(inputs)):
        
        # predict one row of training example
        predict[i] = predict_fun(weights, inputs[i], bias)
        # the error for the whole 'batch', not just for the training example
        error = sum((outputs-predict)**2)
        
        if outputs[i] != predict[i]:
            
            print(f'batch iteration:{iter} outputs:{outputs[i]}, inputs:{inputs[i]} weights:{weights}')
            weights = weights + learning_rate*(outputs[i]*inputs[i])
            
            '''When there is a mistake on +ve (predict -ve(-1) when output is +ve(+1)), 
            equation 1 above will effectively become'''
            # weights = weights + learning_rate*(+1*input) i.e
            # weights = weights + learning_rate*(1*input)

            '''When there is a mistake on -ve (predict +ve(+1) when output is -ve(-1)), 
            equation 1 above will effectively become'''
            # weights = weights + learning_rate*(-1*input) i.e 
            # weights = weights - learning_rate*(1*input) 
            
print(f'batch iteration:{iter} actual:{outputs} predicted:{predict} error:{error} weights:{weights}')
    
```
